```python
import numpy as np # linear algebra operations
import pandas as pd # used for data preparation
import plotly.express as px #used for data visualization
```

```python
#Read the CSV File
netflix=pd.read_csv("/content/netflix_titles.csv")
netflix.head(10)
```

```python
netflix.info()
```

```python
netflix.describe()
```

```python
#Find the Null values in Data set
netflix.isnull().sum()
```

```python
#Fill NAN
netflix['duration']=netflix['duration'].fillna(method= "ffill" )
netflix['rating']=netflix['rating'].fillna(method= "ffill" )
netflix['date_added']=netflix['date_added'].fillna(method= "ffill" )
```

```python
netflix.isnull().sum()
```

```python
#Filtering Indian Content
indian_content = netflix[netflix['country'].str.contains('India', na=False)]
```

```python
#Top 5 Indian Directors
top5Directors = directors.head()
print(top5Directors)
```

```python
#BarChart for the Top 5 Indian Directors
barChart = px.bar(top5Directors, x='Total Count', y = 'Director', title = 'Top 5 Directors on Indian Netflix')
barChart.show()
```

```python
# Most popular genres for Indian content
indian_genres = indian_content['listed_in'].str.split(',', expand=True).stack().to_frame()
indian_genres.columns = ['Genre']
top_indian_genres = indian_genres.groupby('Genre').size().reset_index(name='Total Count').sort_values(by='Total Count', ascending=False).head(10)

```

```python
#Bar Chart for Top Indian Genre
bar_top_indian_genres = px.bar(top_indian_genres, x='Total Count', y='Genre', title='Top 10 Genres in Indian Content')
bar_top_indian_genres.show()
```

```python
# Top actors with the most appearances in Indian content
indian_cast = indian_content['cast'].str.split(',', expand=True).stack().to_frame()
indian_cast.columns = ['Actor']
top_indian_actors = indian_cast.groupby('Actor').size().reset_index(name='Total Count').sort_values(by='Total Count', ascending=False).head(10)

```
