**Indian Netflix Analysis**

```python
import numpy as np # linear algebra operations
import pandas as pd # used for data preparation
import plotly.express as px #used for data visualization
```

```python
#Read the CSV File
df=pd.read_csv("/netflix_titles.csv")
```

```python
#Top 10
df.head(10)
```

```python
#Filter movie content
Movie_content = df[df['type'].str.contains('Movie', na=False)]
```

```python
Movie_content
```

```python
#Filtering Indian Content
indian_content = df[df['country'].str.contains('India', na=False)]
```

```python
indian_content
```

```python
#drop NAN in Data set from director column
indian_content.dropna(subset=['director'])
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
directors_list = directors_list.to_frame()
directors_list.columns = ['Director']
print(directors_list)
```

```python
directors = directors_list.groupby(['Director']).size().reset_index(name='Total Count')
directors = directors.sort_values(by=['Total Count'], ascending = False)
print(directors)
```

```python
#Top 5 Indian Directors
top5Directors = directors.head()
print(top5Directors)
```

```python
#BarChart for the Top 5 Indian Directors
barChart = px.bar(top5Directors, x='Total Count', y = 'Director', title = 'Top 5 Directors on Netflix')
barChart.show()
```

```python
# Most popular genres for Indian content
indian_genres = indian_content['listed_in'].str.split(',', expand=True).stack().to_frame()
indian_genres.columns = ['Genre']
top_indian_genres = indian_genres.groupby('Genre').size().reset_index(name='Total Count').sort_values(by='Total Count', ascending=False).head(10)

```

```python
#Bar Chart for Top Indian Genre
bar_top_indian_genres = px.bar(top_indian_genres, x='Total Count', y='Genre', title='Top 10 Genres in Indian Content')
bar_top_indian_genres.show()
```

```python
# Top actors with the most appearances in Indian content
indian_cast = indian_content['cast'].str.split(',', expand=True).stack().to_frame()
indian_cast.columns = ['Actor']
top_indian_actors = indian_cast.groupby('Actor').size().reset_index(name='Total Count').sort_values(by='Total Count', ascending=False).head(10)
```

```python
#Pie Chart for the top indian Actors
pie_top_indian_actors = px.pie(top_indian_actors, values='Total Count', names='Actor', title='Top 10 Actors in Indian Content')
pie_top_indian_actors.show()
```
